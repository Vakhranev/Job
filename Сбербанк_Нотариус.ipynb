```
import csv, sys, os
from lxml import etree

def main():
    csvFile = 'LC_P_LET_94_13123_38_20200102(1).csv'
    xmlFile = open('myData.xml', 'w')
    csvData = csv.reader(open(csvFile, encoding='utf-8'), delimiter=';')
    ID = []

    header = next(csvData)
    counter = 0
    root = etree.Element('documents')

    for row in csvData:
        if row[78] not in ID:
          ID.append(row[78])
          prod = etree.SubElement(root,'document')
          sbrf = etree.SubElement(prod, 'table')
          for index in range(0, len(header)):
              if index in range(13, 33) or index == 70 or index == 77 or index in range(82, 85) or index == 88 or index in range(90, 95):
                child = etree.SubElement(sbrf, header[index])
                child.text = row[index]
                sbrf.append(child)
              else:
                child = etree.SubElement(prod, header[index])
                child.text = row[index]
              #print(row[index], index)
                prod.append(child)
              #print(prod)
        else:
          sbrf = etree.SubElement(prod, 'table')
          for index in range(0, len(header)):
            if index in range(13, 33) or index == 70 or index == 77 or index in range(82, 85) or index == 88 or index in range(90, 95):
              child = etree.SubElement(sbrf, header[index])
              child.text = row[index]
              sbrf.append(child)

    result = etree.tostring(root, encoding='utf-8', pretty_print=True)
    res = result.decode('utf-8')
    #print(res)
    xmlFile.write(res)


if __name__ == '__main__':
    main()
```
